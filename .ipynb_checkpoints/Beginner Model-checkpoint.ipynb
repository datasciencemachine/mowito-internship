# IMPORTING REQUIRED LIBRARIES

```python
import cv2
import numpy as np
import os
import glob
import matplotlib.pyplot as plt
```

# READING THE TEST IMAGE AND TEMPLATE

```python
img_rgb = cv2.imread('basic/test_image.png')
img_gray = cv2.imread('basic/test_image.png',0)

template = cv2.imread('basic/template_image.png',0)
```

# DETECTING THE TEMPLATE

```python
template = cv2.imread('basic/template_image.png',0)
w, h = template.shape[::-1]
res = cv2.matchTemplate(img_gray,template,cv2.TM_CCOEFF_NORMED)
threshold = 0.9
loc = np.where( res >= threshold)
for pt in zip(*loc[::-1]):
    cv2.rectangle(img_rgb, pt, (pt[0] + w, pt[1] + h), (0,0,255),1)

cv2.imshow('Detected',img_rgb)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.figure(figsize = (100,10))
img = cv2.cvtColor(img_rgb,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

Since all the samples have been detected, therefore opnecv template matching is enough

Here since the template was quite simple, and the template was quite similar to the test image instances, therefore all of the instances were detected easily.
